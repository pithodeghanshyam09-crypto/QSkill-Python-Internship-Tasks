```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Student": ["Amit", "Rahul", "Sneha", "Priya", "Karan", "Neha", "Rohit", "Anjali"],
    "Maths": [78, 85, 90, 72, 65, 88, 76, 95],
    "Science": [82, 79, 92, 70, 68, 91, 74, 96],
    "English": [75, 80, 85, 78, 70, 89, 77, 94]
}

df = pd.DataFrame(data)

df.to_csv("student_data.csv", index=False)

print("CSV file created successfully!")

df = pd.read_csv("student_data.csv")

print("\nFirst 5 Records:")
print(df.head())

print("\nBasic Information:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

print("\nAverage Marks:")
print("Average Maths Marks:", df["Maths"].mean())
print("Average Science Marks:", df["Science"].mean())
print("Average English Marks:", df["English"].mean())

df["Total"] = df["Maths"] + df["Science"] + df["English"]
df["Percentage"] = df["Total"] / 3

print("\nFinal Data:")
print(df)

plt.figure(figsize=(8, 5))
plt.bar(df["Student"], df["Percentage"])
plt.title("Student Percentage Bar Chart")
plt.xlabel("Students")
plt.ylabel("Percentage")
plt.show()

plt.figure(figsize=(8, 5))
plt.scatter(df["Maths"], df["Science"])
plt.title("Maths vs Science Scatter Plot")
plt.xlabel("Maths Marks")
plt.ylabel("Science Marks")
plt.show()

correlation = df[["Maths", "Science", "English", "Total", "Percentage"]].corr()

plt.figure(figsize=(8, 5))
plt.imshow(correlation)
plt.colorbar()
plt.xticks(range(len(correlation)), correlation.columns, rotation=45)
plt.yticks(range(len(correlation)), correlation.columns)
plt.title("Correlation Heatmap")
plt.show()

print("\nInsights:")
print("1. Priya and Karan have comparatively lower marks.")
print("2. Anjali has the highest percentage.")
print("3. Maths and Science marks show a positive relationship.")
print("4. Total and Percentage are strongly correlated.")
```
